```python
import os, json, yaml, random, warnings
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

warnings.filterwarnings("ignore")

# ==========================================================
# ðŸ“‚ PATH SETUP
# ==========================================================
BASE_PATH = r"C:\Users\NXTWAVE\Downloads\Greenhouse Climate & Crop Yield Optimizer"
DATA_PATH = os.path.join(BASE_PATH, r"archive\20210703_greenhouse_data.csv")

MODEL_PATH   = os.path.join(BASE_PATH, "biosphere_model.h5")
SCALER_PATH  = os.path.join(BASE_PATH, "biosphere_scaler.pkl")
CONFIG_PATH  = os.path.join(BASE_PATH, "biosphere_config.yaml")
RESULTS_CSV  = os.path.join(BASE_PATH, "biosphere_results.csv")
RESULTS_JSON = os.path.join(BASE_PATH, "biosphere_prediction.json")

# ==========================================================
# 1ï¸âƒ£ LOAD DATASET
# ==========================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH, sep=";", engine="python", on_bad_lines="skip")
df.columns = [c.strip().replace(" ", "_").lower() for c in df.columns]
df = df.dropna(axis=1, how="all").dropna(axis=0, how="any").reset_index(drop=True)
for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors="ignore")

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) < 2:
    raise ValueError("âŒ Not enough numeric columns for training.")

print(f"[INFO] Numeric columns: {numeric_cols}")

# ==========================================================
# 2ï¸âƒ£ FEATURES / TARGET
# ==========================================================
features = numeric_cols
target = features[-1]
X = df[features[:-1]].values
y = df[target].values.reshape(-1, 1)

# ==========================================================
# 3ï¸âƒ£ SCALING
# ==========================================================
scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

with open(SCALER_PATH, "wb") as f:
    pickle.dump({"x": scaler_x, "y": scaler_y}, f)

# ==========================================================
# 4ï¸âƒ£ CREATE SEQUENCES
# ==========================================================
def create_sequences(X, y, time_steps=5):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

TIME_STEPS = 5
X_seq, y_seq = create_sequences(X_scaled, y_scaled, TIME_STEPS)
train_size = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

# ==========================================================
# 5ï¸âƒ£ HYBRID OPTIMIZATION â€” CSA + ACO
# ==========================================================
def objective(params):
    lr, units, dropout = params
    model = Sequential([
        LSTM(int(units), input_shape=(X_train.shape[1], X_train.shape[2])),
        Dropout(float(dropout)),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss="mse")
    model.fit(X_train, y_train, epochs=4, batch_size=16, verbose=0)
    pred = model.predict(X_test)
    return mean_squared_error(y_test, pred)

def hybrid_csa_aco(num_nests=6, iterations=3):
    # Initialize nests (solutions)
    nests = np.random.rand(num_nests, 3)
    best_nest = nests[0]
    best_score = float("inf")

    for it in range(iterations):
        # --- CSA Step (Levy flight) ---
        for i in range(num_nests):
            step_size = np.random.normal(0, 0.2, 3)
            new_nest = nests[i] + step_size * np.random.randn(3)
            new_nest = np.clip(new_nest, 0, 1)
            lr = 0.0001 + new_nest[0] * 0.0099
            units = int(32 + new_nest[1] * 64)
            dropout = 0.1 + new_nest[2] * 0.4
            score = objective([lr, units, dropout])
            if score < best_score:
                best_score = score
                best_nest = new_nest

        # --- ACO Step (pheromone reinforcement) ---
        pheromones = np.exp(-np.linspace(0, 1, num_nests))
        probs = pheromones / pheromones.sum()
        selected_indices = np.random.choice(range(num_nests), size=num_nests, p=probs)
        new_population = []
        for idx in selected_indices:
            ant = nests[idx] + np.random.uniform(-0.05, 0.05, 3)
            new_population.append(np.clip(ant, 0, 1))
        nests = np.array(new_population)

        print(f"[CSA+ACO] Iter {it+1}/{iterations} â€” Best MSE: {best_score:.6f}")

    lr = 0.0001 + best_nest[0] * 0.0099
    units = int(32 + best_nest[1] * 64)
    dropout = 0.1 + best_nest[2] * 0.4
    return [lr, units, dropout]

print("[INFO] Running Hybrid CSA + ACO Optimization...")
best_lr, best_units, best_dropout = hybrid_csa_aco()
print(f"[INFO] Best Hyperparams â†’ lr={best_lr:.5f}, units={best_units}, dropout={best_dropout:.2f}")

# ==========================================================
# 6ï¸âƒ£ FINAL LSTM TRAINING
# ==========================================================
model = Sequential([
    LSTM(best_units, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(best_dropout),
    Dense(1)
])
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=best_lr),
              loss="mse", metrics=["mae"])
history = model.fit(X_train, y_train, epochs=25, batch_size=16,
                    validation_split=0.1, verbose=1)
model.save(MODEL_PATH)

# ==========================================================
# 7ï¸âƒ£ EVALUATION
# ==========================================================
y_pred = model.predict(X_test)
y_pred_inv = scaler_y.inverse_transform(y_pred)
y_test_inv = scaler_y.inverse_transform(y_test)

mae = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)
print(f"[RESULT] MAE={mae:.4f}, RMSE={rmse:.4f}, RÂ²={r2:.4f}")

results_df = pd.DataFrame({"Actual": y_test_inv.flatten(),
                           "Predicted": y_pred_inv.flatten()})
results_df.to_csv(RESULTS_CSV, index=False)

with open(RESULTS_JSON, "w") as f:
    json.dump({
        "mae": mae, "rmse": rmse, "r2": r2,
        "best_params": {"lr": best_lr, "units": best_units, "dropout": best_dropout}
    }, f, indent=4)

# ==========================================================
# 8ï¸âƒ£ VISUALIZATIONS
# ==========================================================

# Accuracy Graph
plt.figure(figsize=(10, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Training & Validation Loss (Hybrid CSA + ACO)")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_accuracy_graph.png"))

# Heatmap
corr_df = df.select_dtypes(include=[np.number])
if corr_df.shape[1] >= 2:
    plt.figure(figsize=(6, 5))
    sns.heatmap(corr_df.corr(), annot=True, fmt=".2f", cmap="viridis")
    plt.title("Feature Correlation Heatmap")
    plt.tight_layout()
    plt.savefig(os.path.join(BASE_PATH, "biosphere_heatmap.png"))

# Comparison Graph
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv[:200], label="Actual", linewidth=2)
plt.plot(y_pred_inv[:200], label="Predicted", linestyle="dashed")
plt.title("Actual vs Predicted Comparison (Hybrid CSA + ACO)")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_comparison_graph.png"))

# Prediction Graph
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv, color="blue", label="Actual")
plt.plot(y_pred_inv, color="red", linestyle="dotted", label="Predicted")
plt.title("Prediction Trend (Hybrid CSA + ACO)")
plt.legend(); plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_prediction_graph.png"))

# Hybrid CSA + ACO Result Graph
plt.figure(figsize=(6, 6))
sns.scatterplot(x=y_test_inv.flatten(), y=y_pred_inv.flatten(), s=20)
plt.title("Hybrid CSA + ACO Result Graph: Actual vs Predicted")
plt.xlabel("Actual"); plt.ylabel("Predicted")
plt.plot([min(y_test_inv), max(y_test_inv)],
         [min(y_test_inv), max(y_test_inv)], color="red", linestyle="--")
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "CSA_ACO_biosphere_result_graph.png"))
print("[INFO] âœ… Saved CSA_ACO_biosphere_result_graph.png")

# ==========================================================
# 9ï¸âƒ£ CONFIG SAVE
# ==========================================================
config = {
    "optimizer": "Hybrid CSA + ACO",
    "dataset": DATA_PATH,
    "features": features[:-1],
    "target": target,
    "time_steps": TIME_STEPS,
    "best_hyperparams": {
        "learning_rate": float(best_lr),
        "units": int(best_units),
        "dropout": float(best_dropout)
    },
    "metrics": {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2)}
}
with open(CONFIG_PATH, "w") as f:
    yaml.dump(config, f)

print("\n[INFO] âœ… Training complete. Artifacts saved in:")
print(f"ðŸ“¦ {BASE_PATH}")
print("  â”œâ”€â”€ biosphere_model.h5")
print("  â”œâ”€â”€ biosphere_scaler.pkl")
print("  â”œâ”€â”€ biosphere_config.yaml")
print("  â”œâ”€â”€ biosphere_results.csv")
print("  â”œâ”€â”€ biosphere_prediction.json")
print("  â”œâ”€â”€ biosphere_accuracy_graph.png")
print("  â”œâ”€â”€ biosphere_heatmap.png")
print("  â”œâ”€â”€ biosphere_comparison_graph.png")
print("  â”œâ”€â”€ biosphere_prediction_graph.png")
print("  â””â”€â”€ CSA_ACO_biosphere_result_graph.png")
```
