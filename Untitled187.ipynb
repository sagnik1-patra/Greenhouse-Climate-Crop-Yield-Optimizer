```python
import os, json, yaml, random, warnings
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

warnings.filterwarnings("ignore")

# ==========================================================
# 📂 PATH SETUP
# ==========================================================
BASE_PATH = r"C:\Users\NXTWAVE\Downloads\Greenhouse Climate & Crop Yield Optimizer"
DATA_PATH = os.path.join(BASE_PATH, r"archive\20210703_greenhouse_data.csv")

MODEL_PATH   = os.path.join(BASE_PATH, "biosphere_model.h5")
SCALER_PATH  = os.path.join(BASE_PATH, "biosphere_scaler.pkl")
CONFIG_PATH  = os.path.join(BASE_PATH, "biosphere_config.yaml")
RESULTS_CSV  = os.path.join(BASE_PATH, "biosphere_results.csv")
RESULTS_JSON = os.path.join(BASE_PATH, "biosphere_prediction.json")

# ==========================================================
# 1️⃣ LOAD DATASET (SEMICOLON SAFE)
# ==========================================================
print("[INFO] Loading dataset...")

df = pd.read_csv(DATA_PATH, sep=";", engine="python", on_bad_lines="skip")
print("[INFO] ✅ Loaded with semicolon (;) delimiter")
print(f"[INFO] Dataset shape: {df.shape}")

# Clean and preprocess
df.columns = [c.strip().replace(" ", "_").lower() for c in df.columns]
df = df.dropna(axis=1, how="all").dropna(axis=0, how="any").reset_index(drop=True)

for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors="ignore")

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) < 2:
    raise ValueError("❌ Not enough numeric columns for training.")

print(f"[INFO] Numeric Columns: {numeric_cols}")

# ==========================================================
# 2️⃣ FEATURES & TARGET
# ==========================================================
features = numeric_cols
target = features[-1]
X = df[features[:-1]].values
y = df[target].values.reshape(-1, 1)

# ==========================================================
# 3️⃣ SCALING
# ==========================================================
scaler_x, scaler_y = MinMaxScaler(), MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

with open(SCALER_PATH, "wb") as f:
    pickle.dump({"x": scaler_x, "y": scaler_y}, f)

# ==========================================================
# 4️⃣ CREATE LSTM SEQUENCES
# ==========================================================
def create_sequences(X, y, time_steps=10):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

TIME_STEPS = 10
X_seq, y_seq = create_sequences(X_scaled, y_scaled, TIME_STEPS)

train_size = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

# ==========================================================
# 5️⃣ HYBRID OPTIMIZATION (PSO + GWO + HSA)
# ==========================================================
def objective(params):
    lr, units, dropout = params
    model = Sequential([
        LSTM(int(units), input_shape=(X_train.shape[1], X_train.shape[2])),
        Dropout(float(dropout)),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss="mse")
    model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)
    pred = model.predict(X_test)
    return mean_squared_error(y_test, pred)

def pso(num_particles=5, iterations=3):
    best_global, best_score = None, float("inf")
    particles = np.random.rand(num_particles, 3)
    for _ in range(iterations):
        for p in particles:
            lr = 0.0001 + p[0]*0.0099
            units = int(32 + p[1]*128)
            dropout = 0.1 + p[2]*0.4
            score = objective([lr, units, dropout])
            if score < best_score:
                best_score, best_global = score, [lr, units, dropout]
    return best_global

def gwo(base_params):
    lr, units, dropout = base_params
    candidates = []
    for _ in range(3):
        lr_c = lr * random.uniform(0.8, 1.2)
        units_c = units * random.uniform(0.8, 1.2)
        dropout_c = dropout * random.uniform(0.8, 1.2)
        candidates.append([lr_c, int(units_c), dropout_c])
    scores = [objective(c) for c in candidates]
    return candidates[np.argmin(scores)]

def hsa(params):
    lr, units, dropout = params
    lr_h = lr + random.uniform(-0.0005, 0.0005)
    units_h = int(units + random.uniform(-10, 10))
    dropout_h = max(0.1, min(0.5, dropout + random.uniform(-0.05, 0.05)))
    return [lr_h, units_h, dropout_h]

print("[INFO] Running Hybrid Optimization (PSO + GWO + HSA)...")
pso_best = pso()
gwo_best = gwo(pso_best)
final_params = hsa(gwo_best)
best_lr, best_units, best_dropout = final_params
print(f"[INFO] Best Hyperparams → lr={best_lr:.5f}, units={best_units}, dropout={best_dropout:.2f}")

# ==========================================================
# 6️⃣ FINAL LSTM MODEL
# ==========================================================
model = Sequential([
    LSTM(best_units, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(best_dropout),
    Dense(1)
])
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=best_lr),
              loss="mse", metrics=["mae"])

history = model.fit(X_train, y_train, epochs=30, batch_size=32,
                    validation_split=0.1, verbose=1)

model.save(MODEL_PATH)
print(f"[INFO] Model saved → {MODEL_PATH}")

# ==========================================================
# 7️⃣ EVALUATION
# ==========================================================
y_pred = model.predict(X_test)
y_pred_inv = scaler_y.inverse_transform(y_pred)
y_test_inv = scaler_y.inverse_transform(y_test)

mae  = mean_absolute_error(y_test_inv, y_pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2   = r2_score(y_test_inv, y_pred_inv)
print(f"[RESULT] MAE={mae:.4f}, RMSE={rmse:.4f}, R²={r2:.4f}")

results_df = pd.DataFrame({"Actual": y_test_inv.flatten(),
                           "Predicted": y_pred_inv.flatten()})
results_df.to_csv(RESULTS_CSV, index=False)

with open(RESULTS_JSON, "w") as f:
    json.dump({
        "mae": mae, "rmse": rmse, "r2": r2,
        "best_params": {"lr": best_lr, "units": best_units, "dropout": best_dropout}
    }, f, indent=4)

# ==========================================================
# 8️⃣ VISUALIZATIONS (ALL GRAPHS)
# ==========================================================

# (a) Accuracy / Loss Graph
plt.figure(figsize=(10, 5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Training & Validation Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_accuracy_graph.png"))
print("[INFO] ✅ Saved biosphere_accuracy_graph.png")

# (b) Heatmap (numeric columns only)
corr_df = df.select_dtypes(include=[np.number])
if corr_df.shape[1] >= 2:
    plt.figure(figsize=(6, 5))
    sns.heatmap(corr_df.corr(), annot=True, fmt=".2f", cmap="viridis")
    plt.title("Feature Correlation Heatmap")
    plt.tight_layout()
    plt.savefig(os.path.join(BASE_PATH, "biosphere_heatmap.png"))
    print("[INFO] ✅ Saved biosphere_heatmap.png")

# (c) Comparison Graph
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv[:200], label="Actual", linewidth=2)
plt.plot(y_pred_inv[:200], label="Predicted", linestyle="dashed")
plt.title("Actual vs Predicted Comparison (First 200 samples)")
plt.xlabel("Time Index")
plt.ylabel("Target Value")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_comparison_graph.png"))
print("[INFO] ✅ Saved biosphere_comparison_graph.png")

# (d) Prediction Graph (Full Sequence)
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv, color="blue", label="Actual")
plt.plot(y_pred_inv, color="red", linestyle="dotted", label="Predicted")
plt.title("Prediction Trend over Test Data")
plt.xlabel("Samples")
plt.ylabel("Target Value")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_prediction_graph.png"))
print("[INFO] ✅ Saved biosphere_prediction_graph.png")

# (e) Result Graph (Scatter)
plt.figure(figsize=(6, 6))
sns.scatterplot(x=y_test_inv.flatten(), y=y_pred_inv.flatten(), s=20)
plt.title("Result Graph: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.plot([min(y_test_inv), max(y_test_inv)],
         [min(y_test_inv), max(y_test_inv)], color="red", linestyle="--")
plt.tight_layout()
plt.savefig(os.path.join(BASE_PATH, "biosphere_result_graph.png"))
print("[INFO] ✅ Saved biosphere_result_graph.png")

# ==========================================================
# 9️⃣ CONFIG FILE
# ==========================================================
config = {
    "dataset": DATA_PATH,
    "features": features[:-1],
    "target": target,
    "time_steps": TIME_STEPS,
    "optimizer": "Hybrid PSO + GWO + HSA",
    "best_hyperparams": {
        "learning_rate": float(best_lr),
        "units": int(best_units),
        "dropout": float(best_dropout)
    },
    "metrics": {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2)}
}

with open(CONFIG_PATH, "w") as f:
    yaml.dump(config, f)

# ==========================================================
# ✅ COMPLETED
# ==========================================================
print("\n[INFO] ✅ All artifacts & graphs saved successfully in:")
print(f"📦 {BASE_PATH}")
print("  ├── biosphere_model.h5")
print("  ├── biosphere_scaler.pkl")
print("  ├── biosphere_config.yaml")
print("  ├── biosphere_results.csv")
print("  ├── biosphere_prediction.json")
print("  ├── biosphere_accuracy_graph.png")
print("  ├── biosphere_heatmap.png")
print("  ├── biosphere_comparison_graph.png")
print("  ├── biosphere_prediction_graph.png")
print("  └── biosphere_result_graph.png")
```
